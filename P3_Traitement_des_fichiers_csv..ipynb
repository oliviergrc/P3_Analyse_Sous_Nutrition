## Fichier sous-nutrition

```python
data_sous_nutrition = pd.read_csv("sous_nutrition.csv")
```

```python
data_sous_nutrition
```

```python
data_sous_nutrition.Zone.unique()
```

```python
data_sous_nutrition.Année.unique()
```

```python
data_sous_nutrition.Valeur.unique()
# Nous avons 3 choses qui posent problème, les valeurs NaN, le symbole '<' et le type de donnée qui est un objet.
```

```python
data_sous_nutrition.dropna()
# Fonction permettant d'enlever toutes les valeurs NaN de notre tableau.
# On peut le faire ici car la colonne Valeur est très importante et le reste des lignes n'est pas utilisable sans elle.
```

```python
data_sous_nutrition = data_sous_nutrition.dropna()
```

```python
data_sous_nutrition = data_sous_nutrition[data_sous_nutrition.Valeur != '<0.1']
#On enlève cces valeurs car nos populations correspondantes sont inférieures à 100 000 habitants.
```

```python
data_sous_nutrition.Valeur.unique() #vérification, on a bien que des nombres décimaux qui sont proposés.
```

```python
# Nous réitérons la même opération pour les années.
```

```python
data_sous_nutrition['Année'] = data_sous_nutrition['Année'].map({'2012-2014': '2013', '2013-2015': '2014', 
                                                                 '2014-2016': '2015', '2015-2017': '2016', 
                                                                 '2016-2018': '2017', '2017-2019': '2018'})
# Cela nous permet d'inscrire les années médianes, pour faciliter les calculs en joignant les tableaux.
```

```python
data_sous_nutrition #vérification
```

```python
data_sous_nutrition.to_csv("sous_nut2.csv", index=False)
```

## Fichier aide alimentaire.

```python
data_aide_alim
```

```python
data_aide_alim.Valeur.unique()
```

```python
data_aide_alim.Produit.unique()
```

```python
data_aide_alim.Année.unique()
```

```python
data_aide_alim = data_aide_alim.rename(columns={"Pays bénéficiaire" : "Zone"})
# Nous la renommons au cas où nous avons besoin de faire des croisements avec d'autres tableaux.
```

```python
data_aide_alim["Valeur"] = data_aide_alim["Valeur"]/1000
# Afin d'avoir les même unités que le fichier "disponibilité alimentaire", à savoir en milliers de tonnes.
```

```python
data_aide_alim['Année'] = data_aide_alim['Année'].astype(object)
# Nous changeons le type de données pour qu'il soit de type objet, afin de ne  pas le considérer comme un nombre.
```

```python
data_aide_alim # vérification
```

```python
data_aide_alim.to_csv("aide_alim2.csv", index=False)
```

## Fichier disponibilité alimentaire

```python
data_dispo_alim
```

```python
# Comme on peut le voir, on a beaucoup de valeurs NaN, cependant il y a des valeurs qui ne sont pas nulles sur d'autres lignes.
# On va faire le choix de remplir ces valeurs NaN, car les supprimer nous ferait perdre trop de données utilisables.
```

```python
data_dispo_alim.fillna(0) #la fonction 'fillna' nous permet de remplir les NaN par la valeur que l'on souhaite.
```

```python
data_dispo_alim = data_dispo_alim.fillna(0)
```

```python
# On continue d'effectuer les opérations pour voir les valeurs uniques, je ne les affiche pas pour éviter la répétition.
```

```python
data_dispo_alim
```

```python
data_dispo_alim["Disponibilité alimentaire en quantité (kg/personne/an)"]*1000/365
# On cherche à avoir les même unités que pour les autres colonnes de Disponibilité alimentaire.
# On multiplie par 1000 pour convertir en g et on divise par 365 pour convertir en jour.
```

```python
data_dispo_alim["Disponibilité alimentaire en quantité (kg/personne/an)"] = data_dispo_alim["Disponibilité alimentaire en quantité (kg/personne/an)"]*1000/365
```

```python
data_dispo_alim = data_dispo_alim.rename(columns={"Disponibilité alimentaire en quantité (kg/personne/an)" : "Disponibilité alimentaire en quantité (g/personne/jour)"})
# On les renomme pour être en accord avec les nouvelles unités.
```

```python
data_dispo_alim #vérification
```

```python
data_dispo_alim.to_csv("dispo_alim2.csv", index=False)
```
