# Personnes pouvant potentiellement être nourries

```python
data_population
```

```python
data_population['Année'].astype(object)
```

```python
population2017 = data_population[data_population.Année == 2017]
# On veut savoir la totalité de personnes vivantes en 2017.
```

```python
population2017 = population2017.drop(columns='Année')
```

```python
population2017
```

```python
population2017['Valeur'].sum()*1000000
#Nous avons en 2017 au total 7 548 134 111 personnes.
```

```python
dispo = data_dispo_alimentaire.groupby(['Zone']).sum()
# Ici on regroupe toutes les denrées disponibles par pays.
```

```python
dispo
```

```python
calcul = dispo.merge(population2017, how='right', on=['Zone'])
# On va joindre les deux dataframes.
```

```python
calcul.drop(columns=['Aliments pour animaux', 
                     'Autres Utilisations', 'Disponibilité alimentaire en quantité (g/personne/jour)', 
                     'Disponibilité de matière grasse en quantité (g/personne/jour)',
                     'Disponibilité de protéines en quantité (g/personne/jour)',
                     'Disponibilité intérieure', 
                     'Exportations - Quantité', 
                     'Importations - Quantité', 
                     'Nourriture', 
                     'Pertes', 
                     'Production', 
                     'Semences', 'Traitement', 'Variation de stock'],inplace=True )
#On enlève toutes les colonnes qui ne nous servent pas.
# On peut effectuer la même opération simplement en mettant calcul = calcul(['*noms des colonnes que l'on veut garder*'])
```

```python
calcul['Disponibilité alimentaire en Kcal/jour'] = calcul['Disponibilité alimentaire (Kcal/personne/jour)']*calcul['Valeur']*1000000
    # Nous avons pour chaque pays le nombre total de calories disponible par jour.
```

On va diviser le nombre de calories totales par le nombre de calories demandé pour subvenir aux besoin journaliers d'un humain.
En prenant en compte une activité modérée, nous avons 2000 calories pour une femme et 2500 pour un homme.
On a à peu près 50% d'hommes et 50% de femmes à l'échelle mondiale et par pays, à l'exception de pays comme le Qatar.
Nous ferons donc une moyenne à 2250 calories.
source : https://www.futura-sciences.com/sante/questions-reponses/nutrition-calories-femme-homme-besoins-differents-733/

```python
calcul['Personnes pouvant être nourries par jour'] = calcul['Disponibilité alimentaire en Kcal/jour']/2250
```

```python
calcul
```

```python
calcul['Personnes pouvant être nourries par jour'].sum()
# Nous avons donc un total de 9 297 326 501 personnes qui pourraient être nourries avec la totalité de nos ressources.
# C'est plus de 123% de la population mondiale.
```
